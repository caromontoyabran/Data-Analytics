For this notebook I've set initial random values for age, weight, height and size columns. Let's suposed this data is from a baketball team players. 
I customized data by creating 5 interesting graphics. Additionally I attached a special section at the end!!! 
Purpose of the project: just for fun :)

```r
# 1. Set dataset
my_data <- data.frame(
  age = c(20, 22, 23, 25, 26, 20, 26, 23, 20, 30, 28, 25, 27, 25, 28),
  weight = c(80, 75, 90, 85, 80, 70, 60, 55, 60, 60, 70, 75, 85, 90, 85),
  height = c(1.70, 1.60, 1.75, 1.65, 1.80, 1.90, 1.85, 1.95, 1.75, 1.80, 1.90, 1.85, 1.75, 1.90, 1.85),
  size = c("M", "S", "L", "XL", "XL", "L", "L", "XL", "M", "XL", "M", "XL", "XL", "L", "M")
)
```

```r
# Install and load packages 
install.packages("ggplot2")
library(ggplot2)
```

```r
# Create a hexbin plot to see correlation between players height and weight
ggplot(data = my_data, aes(x = height, y = weight)) +
  geom_hex() +
  labs(x = "Height", y = "Weight") +
  ggtitle("Hexbin Plot of Weight vs. Height")

```

```r
# Create a column with the color conditions. For players less than 25 years old we are asigning them to "Baby Players". Take it with humor :)

my_data$color <- ifelse(my_data$age < 25, "Baby Players", "Solo Players")

# Graph the new column into a boxplot chart 
ggplot(data = my_data, aes(y = age, fill = color)) +
    geom_boxplot() +
    scale_fill_manual(name = "Groups", values = c("Baby Players" = "#ADD8E6", "Solo Players" = "black") )
```

```r
library(ggplot2)

# Create intervals for the weight variable
my_data$weight_intervals <- cut(my_data$weight, breaks = 4)  # Divide into 4 intervals

# Define color for each interval 
bin_colors <- c("red", "pink", "gold", "white")  

# Create the histogram with the personalized colors
ggplot(data = my_data) +
  geom_histogram(aes(x = weight, fill = weight_intervals), color = "black", binwidth = 5) +
  scale_fill_manual(values = bin_colors) +
  labs(x = "Player Weight", y = "Count")
```

```r
# Graphic 4

ggplot(data = my_data, aes(x = age, y = weight)) + 
  geom_point(color = "#ed0086", size = 7) + 
  labs(x = "Players' Age", y = "Players' Weight")
```

# Statistical Concepts

In this section I just wanna try some advanced statistics concepts. This is not part of the Players dataset

```r

# Defined the mean and standard deviation using "rnorm" function and store it into a dataframe

the_average <- 300
standard_deviation <- 15
data1 <- rnorm(n = 1000, mean = the_average, sd = standard_deviation)
df <- data.frame(values = data1)
```

```r
# Graphic 5 

ggplot(df, aes(x = values)) +
  geom_density(color = "blue", fill = "lightblue")+
  labs(x = "x", y= "Density")+
  ggtitle("Normal Distribution: Mean = 300, Standard deviation = 15")
```
