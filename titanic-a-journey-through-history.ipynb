## Titanic History

The Titanic was a British luxury passenger liner that tragically sank on its maiden voyage in 1912 after colliding with an iceberg. The disaster resulted in the loss of over 1,500 lives, making it one of the deadliest maritime disasters in history. The Titanic's story has captivated the world and continues to be a subject of fascination.

Survivors' demographics played a significant role in the Titanic's history. This notebook explores some intriguing information about survivors categorized by age, class, and sex.

* Data Source: 'Titanic' dataset provided by R on it's datasets collection
* Project Purpose: complete a basic data analysis process while exploring an interest topic

```r
# Install and load the Titanic dataset 
install.packages('datasets')
library(datasets)
data(Titanic)
```

```r
# Data exploration
str(Titanic)
```

```r

# Check if there are any missing values in the Titanic dataset
any_missing <- any(is.na(Titanic))

# Output the result
if (any_missing) {
  print("The dataset contains missing values.")
} else {
  print("The dataset does not contain missing values.")
}

```

```r
#Install the ggplot2 package and transform the Titanic's dataset into a dataframe

install.packages("ggplot2")
library(ggplot2)
Titanic_df <- as.data.frame(Titanic)

```

### Visual Voyage: Aesthetic Insights

I have created three visually appealing graphs to compare the variables included in the Titanic dataset, namely sex, age, and class.

To ensure accuracy and reliability, it's important to note that these plots were generated using the Titanic dataset provided by R's datasets collection. However, in the event that you encounter any discrepancies or variations in the displayed data, please take them into consideration.

```r

graph <- ggplot(data = Titanic_df) +  
  geom_bar(mapping = aes(x = Sex, y = Freq, fill = Sex), stat = "identity") +
  ggtitle("Comparison of Survivors by Sex") +
  labs(y = "Frequency", caption = "*The label 'Frequency' indicates the rate of each variable described on the graph") +
  scale_fill_manual(values = c("#722F37", "#Efb810")) +
  theme_classic()

print(graph)

```

Conclusion from Graph #1: Here we can see that the most affected by the tragedy were men overall 

```r

graph_1 <- ggplot(data = Titanic_df) +
  geom_bar(mapping = aes(x = Sex, y = Freq, fill = Class), stat = "identity") + 
  ggtitle("Comparison of Survivers by Sex & Class") + labs(y = "Freq", caption = "*The label 'Freq' states for Frequency and it indicates the rate of each variable described on the graph") + theme_classic()
print(graph_1)
```

Conclusion from Graph #2: We can distinct here various statements. Since we had already established that the most afected genre were men. Now we appreciate that fact distributed by classes. For female in the crew we have the lowest indication of deaths. On the other hand for male that were part of the crew we can see the contrarious situation.

```r
graph_2 <- ggplot(data = Titanic_df) + 
  geom_point(mapping = aes(x = Age, y = Freq, shape = Age, size = Freq, alpha = 0.1, color = Sex)) + 
  scale_color_manual(values = c("#722F37", "#Efb810")) + 
  ggtitle("Comparison of Survivors by Age and Sex") + labs(y = "Freq", caption = "*The label 'Freq' states for Frequency and it indicates the rate of each variable described on the graph") + theme_linedraw()
print(graph_2)
```

In this geom_point graph was modified to show an additional figure besides dot, for better interpretation
Conclusion from Graph #3: Men were the most affected in both age ranges.
